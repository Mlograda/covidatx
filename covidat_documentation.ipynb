### What is it?

**Covidat** is a python package that accesses the [gov.uk](https://coronavirus.data.gov.uk/details/developers-guide) Covid-19 data API, downloads selected features and provides simple functions to produce a wide range of well formatted, attractive charts. It enables users with no experience of using APIs or python plotting packages to easily create a range of useful data visualizations, using the latest official data, by implementing a set of functions.  
  
### Why is it needed?

In the midst of a global pandemic, access to data giving insight on case numbers, case rates, deaths, death rates, hospitalizations and vaccine uptake is essential. The UK government has a well maintained and comprehensive API providing a wide range of data covering many aspects of the pandemic. The API is well designed, but interacting with it requires some coding proficiency, which excludes many people from accessing it. This package aims to simplify the work involved in accessing the data and visualizing it.

### How to use this package?

## 1. get_national_data() &#127988;

```python
# Let's retrieve some data using get_national_data()
england = data.CovidData(nation='england').get_national_data()
```

```python
# We now have Covid data for England, let's take a look at it
england.head(10)
```

As you can see, it returns a pandas DataFrame &#128060;. Similar data can be retrieved for Scotland, Northern Ireland and Wales by setting the 'nation' parameter to the country names i.e  
`ni = data.CovidData(nation='northern ireland').get_national_data()`  
`scotland = data.CovidData(nation='scotland').get_national_data()`  
`wales = data.CovidData(nation='wales').get_national_data()`

## 3. get_local_data()

```python
# Let's retrieve some local data
local = data.CovidData().get_local_data(date='2021-12-30')
```

```python
local.head(10)
```

Once again, this returns a pandas DataFrame. You might have noticed this method requires a date, this is due to the amount of data returned by this particular call to the API. This data is very granular and so we have restricted access to just one date at a time. A decision which also ensures we remain within the access limits set by gov.uk for the API

```python
local.shape
```

Currently we do not have a property to print the list of local authorities, however for now this can be done using groupby, eg

```python
local.groupby('name')['case_newDaily'].sum()
```

As is implicit from the above results, unlike the regional data, local data **includes regions from all UK countries**

## 4. get_uk_data()

This method collects data from all nations and sums it to give results for the entire UK

```python
# Let's look at the data for the entire UK
uk_data = data.CovidData().get_uk_data()
```

```python
uk_data.head(10)
```

It returns a summary of the UK data as a pandas DataFrame. As will all the methods except get_local_data(), the data is for the entire duration of the pandemic.

### 1. daily_case_plot()  
This function is to be called on data returned from the get_uk_data() method of CovidData. It will plot the daily covid cases across the UK. It also returns the day with the highest number of Covid cases since the start of the pandemic and the pandemic duration. The optional paramater of save is set to False, if save is set to True the function will save the output to a .png file.

```python
uk_data = data.CovidData().get_uk_data()
plot.daily_case_plot(uk_data)
```

### 2. regional_plot_cases()  
This function collects its own data and returns a geographical plot of case numbers in the 9 UK regions plus Scotland, Wales and Northern Ireland. The optional paramater of save is set to False, if save is set to True the function will save the output to a .png file.

```python
plot.regional_plot_cases()
```

### 3. regional_plot_rate()  
This function collects its own data and will plot the case rate per 100,000 people across the 9 regions of the UK and also in Scotland, Wales and Northern Ireland. The optional paramater of save is set to False, if save is set to True the function will save the output to a .png file.

```python
plot.regional_plot_rate()
```

### 4. heatmap_cases()  
This function requires as input data returned from calling the get_uk_data() method on CovidData(). It will plot a heatmap of the case numbers in the UK.

```python
plot.heatmap_cases(uk_data)
```

### 5. local_rate_plot()  
This function will collect its own data and plot the latest local level rate numbers per 100,000 on a map of the UK. The optional paramater of save is set to False, if save is set to True the function will save the output to a .png file.

```python
plot.local_rate_plot()
```
